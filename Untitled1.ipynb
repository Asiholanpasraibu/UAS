```python
  import numpy as np
  import matplotlib.pyplot as plt

  # Fungsi keanggotaan segitiga yang aman
  def segitiga(x, a, b, c):
      if a == b:
          kiri = 1.0 if x == a else 0.0
      else:
          kiri = (x - a) / (b - a)
      if b == c:
          kanan = 1.0 if x == c else 0.0
      else:
          kanan = (c - x) / (c - b)
      return max(min(kiri, kanan), 0)

  # Fungsi keanggotaan fuzzy
  def derajat_keanggotaan(nama, x):
      return {
          'rendah': segitiga(x, 0, 1.65, 3.3),
          'sedang': segitiga(x, 3.3, 5.0, 6.6),
          'tinggi': segitiga(x, 6.6, 8.3, 10),
      }

  # Fungsi plotting grafik fuzzy
  def plot_grafik(nama_variabel, nilai):
      x = np.linspace(0, 10, 1000)
      rendah = [segitiga(i, 0, 1.65, 3.3) for i in x]
      sedang = [segitiga(i, 3.3, 5.0, 6.6) for i in x]
      tinggi = [segitiga(i, 6.6, 8.3, 10) for i in x]

      plt.figure(figsize=(8, 4))
      plt.plot(x, rendah, label='Rendah')
      plt.plot(x, sedang, label='Sedang')
      plt.plot(x, tinggi, label='Tinggi')
      plt.axvline(nilai, color='black', linestyle='--', label=f'Input: {nilai}')
      plt.title(f'Derajat Keanggotaan - {nama_variabel.capitalize()}')
      plt.xlabel('Nilai')
      plt.ylabel('Derajat Keanggotaan')
      plt.legend()
      plt.grid(True)
      plt.tight_layout()
      plt.show()

  # Fuzzy inference
  def sistem_penilaian(makanan, pelayanan, suasana):
      m_kg = derajat_keanggotaan('makanan', makanan)
      m_pel = derajat_keanggotaan('pelayanan', pelayanan)
      m_suasana = derajat_keanggotaan('suasana', suasana)

      hasil = {'rendah': 0, 'sedang': 0, 'tinggi': 0}
      for km in m_kg:
          for kp in m_pel:
              for ks in m_suasana:
                  nilai = min(m_kg[km], m_pel[kp], m_suasana[ks])
                  if nilai > 0:
                      if 'rendah' in [km, kp, ks]:
                          hasil['rendah'] = max(hasil['rendah'], nilai)
                      elif 'sedang' in [km, kp, ks]:
                          hasil['sedang'] = max(hasil['sedang'], nilai)
                      else:
                          hasil['tinggi'] = max(hasil['tinggi'], nilai)

      kategori_akhir = max(hasil, key=hasil.get)
      return kategori_akhir, hasil

  # === Bagian Input User ===
  print("Silakan masukkan penilaian dari skala 0 - 10:")
  makanan_input = float(input("Kualitas Makanan: "))
  pelayanan_input = float(input("Kualitas Pelayanan: "))
  suasana_input = float(input("Kualitas Suasana Tempat: "))

  # Jalankan logika fuzzy
  kategori, hasil_inferensi = sistem_penilaian(makanan_input, pelayanan_input, suasana_input)
  print(f"\nHasil Penilaian: {kategori.upper()}")
  print("Derajat Inferensi:")
  for k, v in hasil_inferensi.items():
      print(f"  {k}: {v:.2f}")

  # Tampilkan grafik
  plot_grafik("makanan", makanan_input)
  plot_grafik("pelayanan", pelayanan_input)
  plot_grafik("suasana", suasana_input)
```
