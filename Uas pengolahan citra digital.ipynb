```python
# ==========================================================
# APLIKASI PENGOLAHAN CITRA DIGITAL INTERAKTIF
# User bisa memilih operasi untuk mengubah foto
# ==========================================================

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# ----------------------------------------------------------
# 1. Upload Gambar
# ----------------------------------------------------------
print("Silahkan upload gambar...")
uploaded = files.upload()

for filename in uploaded.keys():
    print("File uploaded:", filename)
    img = cv2.imread(filename)
    img_gray = cv2.imread(filename, 0)

# Fungsi untuk menampilkan gambar
def show(img, title="Result"):
    plt.figure(figsize=(5,5))
    if len(img.shape) == 2:
        plt.imshow(img, cmap='gray')
    else:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img)
    plt.title(title)
    plt.axis('off')
    plt.show()

# ----------------------------------------------------------
# 2. Menu Interaktif
# ----------------------------------------------------------
def menu():
    print("\n===== PILIH OPERASI =====")
    print("1. Erosi")
    print("2. Dilasi")
    print("3. Opening")
    print("4. Closing")
    print("5. Morphological Gradient")
    print("6. Blur (Gaussian)")
    print("7. Edge Detection (Canny)")
    print("8. Resize Gambar")
    print("9. Rotate Gambar")
    print("10. Tampilkan ulang gambar asli")
    print("0. Keluar")
    print("==========================")

# Kernel default
kernel = np.ones((5,5), np.uint8)

# ----------------------------------------------------------
# 3. Program Utama
# ----------------------------------------------------------

running = True

while running:
    menu()
    pilihan = input("Masukkan pilihan: ")

    if pilihan == "1":
        result = cv2.erode(img_gray, kernel, iterations=1)
        show(result, "Erosi")

    elif pilihan == "2":
        result = cv2.dilate(img_gray, kernel, iterations=1)
        show(result, "Dilasi")

    elif pilihan == "3":
        result = cv2.morphologyEx(img_gray, cv2.MORPH_OPEN, kernel)
        show(result, "Opening")

    elif pilihan == "4":
        result = cv2.morphologyEx(img_gray, cv2.MORPH_CLOSE, kernel)
        show(result, "Closing")

    elif pilihan == "5":
        result = cv2.morphologyEx(img_gray, cv2.MORPH_GRADIENT, kernel)
        show(result, "Gradient")

    elif pilihan == "6":
        result = cv2.GaussianBlur(img, (7,7), 0)
        show(result, "Gaussian Blur")

    elif pilihan == "7":
        result = cv2.Canny(img_gray, 100, 200)
        show(result, "Edge Detection (Canny)")

    elif pilihan == "8":
        scale = float(input("Masukkan skala (contoh 0.5 untuk mengecilkan): "))
        result = cv2.resize(img, None, fx=scale, fy=scale)
        show(result, f"Resize x{scale}")

    elif pilihan == "9":
        angle = float(input("Masukkan sudut rotasi (derajat): "))
        h, w = img.shape[:2]
        matrix = cv2.getRotationMatrix2D((w/2, h/2), angle, 1)
        result = cv2.warpAffine(img, matrix, (w, h))
        show(result, f"Rotate {angle}")

    elif pilihan == "10":
        show(img, "Gambar Asli")

    elif pilihan == "0":
        print("Program selesai.")
        running = False

    else:
        print("Pilihan tidak valid, coba lagi.")
```
